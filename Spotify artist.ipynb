```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tracks.csv')
```

```python
df.head()
```

```python
# null values

pd.isnull(df).sum()
```

```python
df.info()
```

```python
# find 10 least possible song in this spotify dataframe

sorted_df = df.sort_values('popularity', ascending=True).head(10)
sorted_df
```

```python
df.describe().transpose()
```

```python
# 10 most popular songs with is greater than 90
df_most_popular_90 = df[df['popularity'] > 90].sort_values(by = 'popularity',ascending = False)
df_most_popular_90[:10]
```

```python
# 10 most popular songs with is greater than 90

most_popular = df.query('popularity > 90', inplace=False).sort_values('popularity',ascending=False)
most_popular[:10]
```

```python
# setting index to release data in main data frame

df.set_index('release_date',inplace = True)
df.index=pd.to_datetime(df.index)
df
```

```python
df[["artists"]].iloc[18]
```

```python
# convert the mili_seconds duration into seconds.
# 1 seconds = 1000 mili_seconds
df['duration_sec'] = round(df['duration_ms']/1000)
# or ------->>>>
#
# df['duration_sec'] = df['duration_ms'].apply(lambda x: round(x/1000))
#
df.drop('duration_ms',inplace = True, axis=1)
df.head(5)
```

```python
# drop ['key','mode','explicit'] and plot corr heat map 
plt.figure(figsize=(12,8))
corr_df = df.drop(['key','mode','explicit'],axis = 1).corr(method='pearson')
heat_map = sns.heatmap(corr_df, annot= True, cmap='inferno', linewidth=1,linecolor='Black')
heat_map.set_title("Correlation HeatMap Between Variables")
heat_map.set_xticklabels(heat_map.get_xticklabels(),rotation = 90)
```

```python
sample_df = df.sample(int(0.004*len(df)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df,y="loudness",x="energy", color="c").set(title = "loudness vs Energy")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df,y="popularity",x="acousticness", color="b").set(title = "Popularity vs Acousticness")
```

```python
# create year column from release date 
df.head()
```

```python
# get the index values and make a seperate column
# convert the value into datetime using pd.datetime
# extract the year using dt.year
# next plot a histogram 
df['year'] = df.index.get_level_values('release_date')
df['year'] = pd.to_datetime(df['year'],dayfirst = True).dt.year
df.head()
plt.figure(figsize=(12,10))
sns.displot(df['year'],discrete = True,aspect = 2, height= 5, kind = 'hist').set(title = 'Number of songs per year');
```

```python
df.head()
```

```python
#plot duration vs year 
plt.figure(figsize=(30,15))

sns.barplot(x = df.year, y = df.duration_sec,errwidth = False).set(title='')
plt.xticks(rotation=45);
```
